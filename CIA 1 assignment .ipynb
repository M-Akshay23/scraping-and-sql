```python
#create a connection to the daatbase
conn=sqlite3.connect('textile.db')
#create a cursor
c=conn.cursor()
```

```python
# create table
c.execute("""
CREATE TABLE IF NOT EXISTS textile_finance (
    'S.No.' INT, 'Name' TEXT, 'CMP  Rs.' FLOAT, 'P/E' FLOAT, 'Mar Cap  Rs.Cr.' FLOAT, 'Div Yld  %' FLOAT,
       'NP Qtr  Rs.Cr.' FLOAT, 'Qtr Profit Var  %' FLOAT, 'Sales Qtr  Rs.Cr.' FLOAT,
       'Qtr Sales Var  %' FLOAT, 'ROCE  %' FLOAT, 'ROE  %' FLOAT, 'Down  %' );
""")
```

```python
df.to_sql('textile_finance', conn, if_exists='replace', index=False)
```

```python
#create a wrapper function
def rsql(query):
    return pd.read_sql_query(query,conn)
```

```python
rsql("""
SELECT * FROM textile_finance
WHERE "P/E" > 40
""")
```

```python
rsql("""
SELECT "ROE %", COUNT(*) AS Count
FROM textile_finance
GROUP BY "ROE %"
""")
```

```python
rsql("""
SELECT "Name", "CMP Rs." 
FROM textile_finance
ORDER BY "CMP Rs." DESC
LIMIT 10
""")
```

```python
rsql("""
SELECT * FROM textile_finance
LIMIT 5
""")
```

```python
rsql("""
SELECT "Name" AS Company, "CMP Rs." AS Price
FROM textile_finance
""")
```

```python
rsql("""
SELECT "Div Yld %", COUNT(*) AS Count
FROM textile_finance
GROUP BY "Div Yld %"
HAVING Count > 1
""")
```

```python
rsql("""
SELECT "Name", "P/E"
FROM textile_finance
WHERE "P/E" BETWEEN 15 AND 30
""")
```

```python
rsql("""
SELECT 
  AVG("P/E") AS "Avg_P_E",
  SUM("Sales Qtr Rs.Cr.") AS Total_Sales,
  COUNT(*) AS Company_Count
FROM textile_finance
""")
```
